# XGBoost Algorithm

The XGBoost or Extreme Gradient Boosting algorithm is a decision tree based machine learning algorithm 
which uses a process called boosting to help improve performance.

```python
from sklearn.model_selection import train_test_split
```

```python
#train_test_split
x_train, x_test, y_train, y_test = train_test_split(features,labels, test_size = 0.30, random_state = 42)
```

```python
x_train.shape, x_test.shape, y_train.shape,y_test.shape
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
xgb=XGBClassifier(n_estimators=100,random_state=0,learning_rate=0.2)
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
y_pred
```

```python
y_pred.shape
```

```python
#predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy :", accuracy )
```
