# KNN (K Nearest Neighbor) Algorithm

The abbreviation KNN stands for “K-Nearest Neighbor”. It is a supervised machine learning algorithm.
The algorithm can be used to solve both classification and regression problem statements. The number
of nearest neighbours to a new unknown variable that has to be predicted or classified is denoted by the symbol 'K'.
